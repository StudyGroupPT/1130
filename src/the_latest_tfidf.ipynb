```
# ライブラリをインポートする
from math import log
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
# 正規表現
def regex(txtfile):
  txtfile = txtfile.split()
  txtfile = [item.replace(",", "").replace("(", "").replace(")", "").replace("!", "").replace(".", "").lower() for item in txtfile]
  return txtfile

def fileread(file):
  f = open(file)
  filename = f.read()
  f.close()
  filename = regex(filename)
  return filename
```

```
    
# beatlesの歌詞を読み込んでそれぞれ正規表現する。
let_it_be = fileread('let_it_be.txt')
dig_a_pony = fileread('dig_a_pony.txt')
dig_it = fileread('dig_it.txt')
for_you_blue = fileread('for_you_blue.txt')
get_back = fileread('get_back.txt')
I_me_mine = fileread('I_me_mine.txt')
ive_got_a_feeling = fileread('ive_got_a_feeling.txt')
maggie_mae = fileread('maggie_mae.txt')
one_after_909 = fileread('one_after_909.txt')
the_long_and_winding_road = fileread('the_long_and_winding_road.txt')
two_of_us = fileread('two_of_us.txt')

# 単語を全部足す
beatles = let_it_be + dig_a_pony + dig_it + for_you_blue\
+ get_back + I_me_mine + maggie_mae + one_after_909 + the_long_and_winding_road + two_of_us
```

```
# 読み込み対象文章を配列化する
docs = np.array([
        let_it_be,
        dig_a_pony,
        dig_it,
        for_you_blue,
        get_back,
        I_me_mine,
        maggie_mae,
        one_after_909,
        the_long_and_winding_road,
        two_of_us,
        ])

# 文章に名前をつける
name = np.array([
        'let it be',
        'dig a pony',
        'dig it',
        'for you blue',
        'get back',
        'I me mine',
        'maggie mae',
        'one after 909',
        'the long and winding road',
        'two of_us',
        ])
```

```
# 重複排除
beatles_no_duplication = []
for x in beatles:
    if x not in beatles_no_duplication:
        beatles_no_duplication.append(x)
        


# 文章数を変数Nに代入する
N = len(docs)

# tf計算 単語/1つの文章に出てくる単語の総数 (単語の出現頻度)
def tf(t, d):
    return d.count(t)/len(d)

# idf計算　全文書数/単語tを含む文書数 (各単語のレア度)
def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    
    return log(N/df)+1

# tfidf計算　(単語の出現頻度)∗(各単語のレア度)
def tfidf(t, d):
  return tf(t,d)* idf(t)
```

```
result = []

# 文章ごとにtf関数に重複なしの単語と文章を渡す
for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(beatles_no_duplication)):
        t = beatles[j]
        result[-1].append(tf(t,d))

# 計算結果とカラム名をDataFrameに引数で渡す
tf_ = pd.DataFrame(result, columns=beatles_no_duplication,index=name)

# tfの計算結果をいったん出力
tf_
```

```
result = []
#idf関数に単語を一つずつ渡す
for j in range(len(beatles_no_duplication)):
    t = beatles_no_duplication[j]
    result.append(idf(t))

# 計算結果とカラム名をDataFrameに引数で渡す
idf_ = pd.DataFrame(result, index=beatles_no_duplication, columns=["IDF"])
idf_
```

```
result = []
# 1つの文章　対　全文章の全単語をtfidf関数に渡す
for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(beatles_no_duplication)):
        t = beatles_no_duplication[j]
        
        result[-1].append(tfidf(t,d))
```

```
tfidf_ = pd.DataFrame(result, columns=beatles_no_duplication,index=name)
tfidf_.loc['let it be'].sort_values(ascending=False).head(10)
```
